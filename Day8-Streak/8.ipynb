```python
# get the data for Mumbai Location where the sales GE 300. 
# Get only StoreName, TotalCost, OperatingCost
```

```python
# this is a two step process, NOT very much recommended
stores[ (stores.Location == 'Mumbai') & (stores.TotalSales >= 300) ][['StoreName', 'TotalSales', 'OperatingCost']]
```

```python
# 1 step process, should be used for practical purpose
stores.loc[ (stores.Location == 'Mumbai') & (stores.TotalSales >= 300), 
                       ['StoreName', 'TotalSales', 'OperatingCost']]
```
