```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
var_random_data = pd.Series( np.random.randint( 1000, 10000, 50000 ) )
var_normal_data = pd.Series( np.random.normal( loc = 110, scale = 30, size = 50000) )
#in this --loc = 110, scale = 30, size = 50000) - loc is mean , scale is standard deviation , size is observation.
```

```python
var_normal_data
```

```python
sns.distplot( var_random_data )
plt.show()
```

```python
var = var_random_data
list_means = []
iR = 50
iC = 5

print('Distribution Chart of the original variable - population')
sns.distplot( var )
plt.show()

print('Charts for the means of the samples')
fig, axis = plt.subplots( iR, iC, figsize = (iC * 3, iR * 2))

for iRows in range(iR):
    for iCols in range(iC):
        
        # selecting the sample of 250 obs randomly and adding the mean of that to the list 
        list_means.append(  var.iloc[ np.random.randint(0, var.shape[0], iR * iC) ].mean()  )
        
        # plot the means of the samples
        sns.distplot( pd.Series(list_means), bins = 10, ax = axis[iRows, iCols] )
        
fig.tight_layout()
```
