#### percentile method

```python
p0 = stores.TotalSales.quantile(0)
p100 = stores.TotalSales.quantile(1)

p1 = stores.TotalSales.quantile(0.01)
p99 = stores.TotalSales.quantile(0.99)
```

```python
print('min:', p0, '| lc:', p1, '| uc:', p99, '| max:', p100)
```

#### IQR method

```python
p0 = stores.TotalSales.quantile(0)
p100 = stores.TotalSales.quantile(1)

p25 = stores.TotalSales.quantile(0.25)
p75 = stores.TotalSales.quantile(0.75)

IQR = p75 - p25

lc = p25 - 1.5 * IQR
uc = p75 + 1.5 * IQR

print('min:', p0, '| lc:', lc, '| uc:', uc, '| max:', p100)
```

```python
stores.TotalSales.plot(kind = 'box')
```

```python
p0 = stores.Total_Customers.quantile(0)
p100 = stores.Total_Customers.quantile(1)

p25 = stores.Total_Customers.quantile(0.25)
p75 = stores.Total_Customers.quantile(0.75)

IQR = p75 - p25

lc = p25 - 1.5 * IQR
uc = p75 + 1.5 * IQR

print('min:', p0, '| lc:', lc, '| uc:', uc, '| max:', p100)
```

```python
stores.Total_Customers.plot(kind = 'box')
```

```python
stores.TotalSales.clip( lower = lc, upper = uc ).head(2)
```
